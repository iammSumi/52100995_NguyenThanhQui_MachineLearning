# Import thư viện

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
```

# Đọc dữ liệu từ file CSV


```python
data = pd.read_csv('data.csv') # 1000 dòng đầu
```

# Tiền xử lý dữ liệu

```python
X = data.drop('NObeyesdad', axis=1)
y = data['NObeyesdad']
```

# Xử lý biến categorical

```python
X = pd.get_dummies(X)
```

# Chuẩn hóa biến numeric

```python
scaler = StandardScaler()
X[['Age', 'Height', 'Weight', 'NCP', 'FAF', 'TUE']] = scaler.fit_transform(X[['Age', 'Height', 'Weight', 'NCP', 'FAF', 'TUE']])
```

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Huấn luyện mô hình ban đầu

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

# Kiểm tra mô hình ban đầu

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

# Continual Learning - Tái huấn luyện mô hình khi có dữ liệu mới

```python
new_data = pd.read_csv('data_new.csv')  # Đọc dữ liệu mới từ file CSV 1000 data mới
```

# Tiền xử lý dữ liệu mới

```python
X_new = new_data.drop('NObeyesdad', axis=1)
y_new = new_data['NObeyesdad']
X_new = pd.get_dummies(X_new)
X_new[['Age', 'Height', 'Weight', 'NCP', 'FAF', 'TUE']] = scaler.transform(X_new[['Age', 'Height', 'Weight', 'NCP', 'FAF', 'TUE']])
```

# Tái huấn luyện mô hình với dữ liệu mới

```python
model.fit(X_new, y_new)
```

# Test Production - Đánh giá mô hình trong môi trường triển khai


```python
y_pred_new = model.predict(X_new)
print(classification_report(y_new, y_pred_new))
```
